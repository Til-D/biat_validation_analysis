# Agreement Rates 
Vatavu et al., 2015: https://dl.acm.org/citation.cfm?id=2702223

```python
def agreement_rate(attribute_dict):
    numerator = 0
    total = 0
    for attr in attribute_dict:
        count = attribute_dict[attr]
        numerator += (count * (count-1))/2
        total += count

    return numerator / ((total * (total-1)/2))

def combine_batches(b1, b2):
    batches = deepcopy(b1)

    for key in b2:
        if key not in batches:
            print('ERROR: cannot combine batches with different keys.')
            return {}
        else:
            words_b1 = batches[key]
            words_b2 = b2[key]
            
            overlapping_words = words_b1.keys() & words_b2.keys()
            for word in words_b2:
                if word in overlapping_words:
                    batches[key][word] = words_b1[word] + words_b2[word]
                else:
                    batches[key][word] = words_b2[word]
    return batches

'''
    helper function: returns highest counted word
'''
def get_max_key(wordlist):
    k = ''
    c = 0
    for key in wordlist:
        if wordlist[key] > c:
            k = key
            c = wordlist[key]
    return k

'''
    returns a list of words desc
'''
def get_max_keys(wordlist):
    result = []
    l = deepcopy(wordlist)
    while len(l) > 0:
        key = get_max_key(l)
        result.append(key)
        del l[key]
    return result

'''
    removes all n/a entries from batch
'''
def removeNAs(batch):
    batch = deepcopy(batch)
    for key in batch:
        if 'n/a' in batch[key]:
            del batch[key]['n/a']
    return batch

batches_combined = combine_batches(BATCH_1, BATCH_2)
batches_combined_na = removeNAs(combine_batches(BATCH_1, BATCH_2))

# batch = BATCH_1
batch1 = removeNAs(BATCH_1)
batch1_na = BATCH_1
batch2 = removeNAs(BATCH_2)
batch2_na = BATCH_2

for key in batch1:
    print('Agreement rates', key, ':')
    print('batch_1: %f'% (agreement_rate(batch1[key])))
    print('>', batch1[key])
    print('batch_1_na: %f'% (agreement_rate(batch1_na[key])))
    print('>', batch1_na[key])
    print('batch_2: %f' % (agreement_rate(batch2[key])), batch2[key])
    print('>', batch2[key])
    print('batch_2_na: %f' % (agreement_rate(batch2_na[key])), batch2_na[key])
    print('>', batch2_na[key])
    print('combined: %f'% (agreement_rate(batches_combined[key])), batches_combined[key])
    print('>', batches_combined[key])
    print('combined_na: %f'% (agreement_rate(batches_combined_na[key])), batches_combined_na[key])
    print('>', batches_combined_na[key])
    print('')
```

**Legend:**
- Agreement rate of 0 -> no agreement whatsoever
- Agreement rate of 1 -> perfect agreement
- combined: both batches
- batch_x: all suggestions in batch x without taking into account participants who did not provide any word suggestions ('n/a')
- barch_x_na: all suggestions in batch x while taking into account participants who did not provide any word suggestions ('n/a')


# Disagreement Rates 
Vatavu et al., 2015: https://dl.acm.org/citation.cfm?id=2702223

```python
def disgreement_rate(attribute_dict):
    numerator = 0
    total = 0
    total = sum([attribute_dict[key] for key in attribute_dict])
    
    for attr in attribute_dict:
        count = attribute_dict[attr]
        numerator += (count * (total-count))/2

    return numerator / ((total * (total-1)/2))

for key in batch1:
    print('Disgreement rates', key, ':')
    print('batch_1: %f'% (disgreement_rate(batch1[key])))
    print('>', batch1[key])
    print('batch_1_na: %f'% (disgreement_rate(batch1_na[key])))
    print('>', batch1_na[key])
    print('batch_2: %f' % (disgreement_rate(batch2[key])), batch2[key])
    print('>', batch2[key])
    print('batch_2_na: %f' % (disgreement_rate(batch2_na[key])), batch2_na[key])
    print('>', batch2_na[key])
    print('combined: %f'% (disgreement_rate(batches_combined[key])), batches_combined[key])
    print('>', batches_combined[key])
    print('combined_na: %f'% (disgreement_rate(batches_combined_na[key])), batches_combined_na[key])
    print('>', batches_combined_na[key])
    print('')
```
